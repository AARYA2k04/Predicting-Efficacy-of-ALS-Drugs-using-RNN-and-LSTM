```python
import pandas as pd
from rdkit import Chem
import networkx as nx

# Load the dataset
df = pd.read_csv("ec50.csv")

# Function to convert SMILES to graph and extract features
def smiles_to_graph_and_features(smiles):
    # Load molecule from SMILES
    molecule = Chem.MolFromSmiles(smiles)
    if molecule is None:
        return None, None, None

    # Create an empty graph
    graph = nx.Graph()

    # Initialize dictionaries to store node and edge features
    node_features = {}
    edge_features = {}

    # Add nodes for atoms
    for atom in molecule.GetAtoms():
        atom_idx = atom.GetIdx()
        graph.add_node(atom_idx)
        # Add node features
        node_features[atom_idx] = {
            'atomic_number': atom.GetAtomicNum(),
            'formal_charge': atom.GetFormalCharge(),
            'hybridization': atom.GetHybridization(),
            'num_hydrogens': atom.GetTotalNumHs(),

            # Add more features as needed
        }

    # Add edges for bonds
    for bond in molecule.GetBonds():
        begin_atom_idx = bond.GetBeginAtomIdx()
        end_atom_idx = bond.GetEndAtomIdx()
        # Add edge features
        edge_features[(begin_atom_idx, end_atom_idx)] = {
            'bond_type': bond.GetBondType(),
            # Add more features as needed
        }
        graph.add_edge(begin_atom_idx, end_atom_idx)

    return graph, node_features, edge_features

# Convert SMILES to graphs and extract features
graphs = []
node_features_list = []
edge_features_list = []
for smiles in df['canonical_smiles']:
    graph, node_features, edge_features = smiles_to_graph_and_features(smiles)
    if graph is not None:
        graphs.append(graph)
        node_features_list.append(node_features)
        edge_features_list.append(edge_features)
print(node_features_list)
print(edge_features_list)
# Now 'graphs' is a list of NetworkX graphs representing molecules
# 'node_features_list' contains node features for each molecule
# 'edge_features_list' contains edge features for each molecule
```
